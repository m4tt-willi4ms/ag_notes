```python
from sympy import I, exp, pi, symbols

z = symbols("z")

e = exp(2 * pi * I / 5)
f = z * (z**10 + 11 * z**5 - 1)
H = z**30 + 522 * (z**25 - z**5) - 10005 * (z**20 - z**10) + 1
n = e**2 + e**3
m = e + e**4
O1 = (z**2 + 1) * (z**2 - 2 * n * z - 1) * (z**2 - 2 * m * z - 1)
x1 = -(f**2) / H * O1
```

```python
B = -(f**5) / H**2
B
```

```python
theta = symbols("theta")


def eval_funcs(val):
    return [x.subs(z, val).evalf() for x in [B, O1, x1]]


eval_funcs(exp(2 * pi * I * (1 - 0.0001)))
```
